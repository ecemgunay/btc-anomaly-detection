This mini project demonstrates how to identify statistically significant anomalies in Bitcoin closing prices using the Z-score method. The analysis is done using Python and basic data science libraries, with a clear and visual explanation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the price data (make sure the CSV is in the /data folder)
df = pd.read_csv("data/btc_price_data.csv", parse_dates=["Date"], index_col="Date")
df.head()
```

```python
df['Z-Score'] = (df['Price'] - df['Price'].mean()) / df['Price'].std()
df['Anomaly'] = np.abs(df['Z-Score']) > 1.5
```

```python
anomalies = df[df['Anomaly']]

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Price'], label='BTC/USDT Price', color='blue')
plt.scatter(anomalies.index, anomalies['Price'], color='red', label='Anomaly', zorder=5)
plt.title('Anomaly Detection in BTC Prices (Z-Score > 1.5)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## ðŸ§  Conclusion

By using a simple Z-score method, we were able to detect statistically unusual changes in Bitcoin prices. While this is not a predictive model, such anomaly detection can be useful for triggering alerts, identifying market shocks, or supporting financial risk management in crypto platforms.

This project demonstrates how even simple statistical methods can produce valuable insights when applied to real-world data.
